# Pencile Sketch

Importaing Library

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

# Performing CV

Reading Image

```python
image = mpimg.imread('F:\Oasis\lets\dog.png')
plt.imshow(image)
```

Grey Image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

Inverted Grey Image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image, cmap='Greys_r')
```

Blurred Image

```python
inverted_blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('inverted gray')
ax1.imshow(inverted_image, cmap='Greys_r')

ax2.set_title('inverted blurred image')
ax2.imshow(inverted_blurred, cmap='Greys_r')
```

Pencil Sketch

```python
final = cv2.divide(gray_image, 255 - inverted_blurred, scale=240)
plt.figure(figsize=(5,5))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
